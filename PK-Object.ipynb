# Properti objek
Setiap objek memiliki **identitas**, **tipe**, dan **nilai**:

**Identitas objek** adalah bilangan bulat unik yang dikaitkan dengan objek tersebut. Umumnya, bilangan bulat ini merujuk ke lokasi memori tempat objek disimpan. Setelah dibuat, identitas objek tidak pernah berubah. Fungsi `id()` mengembalikan identitas objek.

**Tipe objek** menentukan kemungkinan nilai dan operasi objek. Misalnya: Bilangan bulat dan float dapat "dibagi" menggunakan operator /, tetapi string tidak bisa. Fungsi `type()` bawaan mengembalikan tipe objek.

**Nilai objek** mewakili status objek saat ini. Banyak objek, seperti angka dan string, tidak dapat dimodifikasi setelah dibuat. Beberapa objek, seperti list (diperkenalkan kemudian), dirancang untuk dimodifikasi.
